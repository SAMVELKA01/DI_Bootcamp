```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats

# Charger le dataset
df = pd.read_csv("AAPL.csv")

# Afficher les 5 premières lignes
print(df.head())
```

```python
# Informations générales
print(df.info())
```

```python
# Supprimer la ligne contenant "Date"
df = df.iloc[1:].copy()

# Renommer la colonne Price en Date
df.rename(columns={"Price": "Date"}, inplace=True)

# Conversion de la date
df["Date"] = pd.to_datetime(df["Date"])

# Conversion des colonnes numériques
colonnes = ["Close", "High", "Low", "Open", "Volume"]

for col in colonnes:
    df[col] = pd.to_numeric(df[col])

# Définir la date comme index
df.set_index("Date", inplace=True)

print(df.head())
print(df.info())
```

```python
# Résumé statistique
print(df.describe())
```

```python
plt.figure(figsize=(12,6))

plt.plot(df.index,
         df["Close"],
         label="Close Price")

plt.title("Apple Closing Price Over Time")
plt.xlabel("Date")
plt.ylabel("Price ($)")
plt.legend()

plt.show()
```

```python
plt.figure(figsize=(12,6))

plt.plot(df.index,
         df["Volume"])

plt.title("Trading Volume")
plt.xlabel("Date")
plt.ylabel("Volume")

plt.show()
```

```python
import mplfinance as mpf

mpf.plot(
    df.tail(200),
    type='candle',
    volume=True,
    style='yahoo',
    title='Apple Candlestick Chart'
)
```

```python
mean_close = np.mean(df["Close"])
print("Mean:", mean_close)
```

```python
median_close = np.median(df["Close"])
print("Median:", median_close)
```

```python
std_close = np.std(df["Close"])

print("Standard Deviation:", std_close)
```

```python
df["Daily_Return"] = df["Close"].pct_change()

print(df["Daily_Return"].head())
```

```python
df["MA30"] = df["Close"].rolling(window=30).mean()
```

```python
plt.figure(figsize=(14,6))

plt.plot(df["Close"], label="Close Price")
plt.plot(df["MA30"], label="MA30")

plt.legend()
plt.title("Closing Price vs Moving Average")

plt.show()
```

```python
year_2022 = df[df.index.year == 2022]["Close"]

year_2023 = df[df.index.year == 2023]["Close"]
```

```python
t_stat, p_value = stats.ttest_ind(
    year_2022,
    year_2023,
    equal_var=False
)

print("T-statistic:", t_stat)
print("P-value:", p_value)
```

```python
plt.figure(figsize=(10,5))

plt.hist(
    df["Daily_Return"].dropna(),
    bins=50
)

plt.title("Distribution of Daily Returns")

plt.show()
```

```python
sample = df["Daily_Return"].dropna().sample(500)

stat, p = stats.shapiro(sample)

print("Statistic:", stat)
print("P-value:", p)
```

```python
moving_avg = np.convolve(
    df["Close"],
    np.ones(30)/30,
    mode="valid"
)
```

```python
df["MA50"] = df["Close"].rolling(50).mean()
df["MA100"] = df["Close"].rolling(100).mean()

clean = df[["MA50", "MA100"]].dropna()

corr = np.corrcoef(clean["MA50"], clean["MA100"])

print(corr)
```

```python
corr_price_volume = df["Close"].corr(df["Volume"])

print("Correlation:", corr_price_volume)
```

```python
from scipy.signal import savgol_filter

df["Smoothed_Close"] = savgol_filter(
    df["Close"],
    window_length=21,
    polyorder=2
)
```

```python
plt.figure(figsize=(12,6))

plt.plot(
    df["Close"],
    alpha=0.5,
    label="Original"
)

plt.plot(
    df["Smoothed_Close"],
    linewidth=3,
    label="Smoothed"
)

plt.legend()
plt.show()
```

L'analyse statistique des actions Apple révèle une croissance globale du cours sur le long terme malgré plusieurs périodes de forte volatilité. Les moyennes mobiles ont permis d'identifier clairement les tendances principales. Les tests statistiques montrent que les prix moyens diffèrent significativement entre certaines années. L'étude des rendements quotidiens indique une distribution non parfaitement normale, caractéristique classique des marchés financiers. Les analyses de corrélation montrent également une relation notable entre plusieurs indicateurs techniques.
