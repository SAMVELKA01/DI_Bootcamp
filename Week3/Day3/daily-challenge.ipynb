```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# Charger le dataset
df = pd.read_csv("mobile_train.csv")

# Aperçu
print(df.head())
```

```python
print(df.info())
print(df.describe())
print(df.isnull().sum())
```

```python
missing = df.isnull().sum()
print(missing)
```

```python
# Exemple si colonne catégorielle existe
df = pd.get_dummies(df, drop_first=True)
```

```python
print(df.head())
```

```python
from scipy import stats

print("Mean RAM:", np.mean(df["ram"]))
print("Median RAM:", np.median(df["ram"]))
print("Mode RAM:", stats.mode(df["ram"], keepdims=True))
```

```python
print("Variance RAM:", np.var(df["ram"]))
print("Std RAM:", np.std(df["ram"]))
print("Range RAM:", np.max(df["ram"]) - np.min(df["ram"]))
```

```python
print("Skewness:", stats.skew(df["battery_power"]))
print("Kurtosis:", stats.kurtosis(df["battery_power"]))
```

```python
corr = df.corr()

print(corr["price_range"].sort_values(ascending=False))
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(df.corr(), cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
df.hist(figsize=(15,10), bins=20)
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x="price_range", y="ram", data=df)
plt.title("RAM vs Price Range")
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.scatter(df["battery_power"], df["price_range"])
plt.xlabel("Battery Power")
plt.ylabel("Price Range")
plt.title("Battery vs Price")
plt.show()
```

```python
group0 = df[df["price_range"] == 0]["ram"]
group3 = df[df["price_range"] == 3]["ram"]

t_stat, p_value = stats.ttest_ind(group0, group3)

print("T-stat:", t_stat)
print("P-value:", p_value)
```

- La RAM est le facteur le plus déterminant du prix.
- La batterie et la mémoire interne ont aussi un impact fort.
- Les variables comme "blue", "dual_sim" ont peu d'influence.
- Les classes de prix sont bien séparées statistiquement.
