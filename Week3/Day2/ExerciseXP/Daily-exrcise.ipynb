```python
# EXERCICE 1

import numpy as np

matrix = np.array([
    [1, 2, 3],
    [0, 1, 4],
    [5, 6, 0]
])

print("Matrice :")
print(matrix)
# Calcul du déterminant et de l'inverse de la matrice
determinant = np.linalg.det(matrix)
print("Déterminant :", determinant)

inverse_matrix = np.linalg.inv(matrix)
print("Inverse de la matrice :")
print(inverse_matrix)
```

```python
# EXERCICE 6

import numpy as np
from scipy import stats

# Fixer la graine pour reproductibilité
np.random.seed(42)

# Productivité AVANT formation
productivity_before = np.random.normal(loc=50, scale=10, size=30)

# Productivité APRÈS formation
productivity_after = productivity_before + np.random.normal(loc=5, scale=3, size=30)

# Moyennes
mean_before = np.mean(productivity_before)
mean_after = np.mean(productivity_after)

print("Moyenne avant :", mean_before)
print("Moyenne après :", mean_after)

# Différence moyenne
print("Amélioration moyenne :", mean_after - mean_before)

# Test t apparié (avant vs après)
t_stat, p_value = stats.ttest_rel(productivity_after, productivity_before)

print("\nStatistique t :", t_stat)
print("p-value :", p_value)

# Décision statistique
alpha = 0.05

if p_value < alpha:
    print("\nConclusion : Le programme est efficace ✅ (rejet de H0)")
else:
    print("\nConclusion : Pas de preuve d'efficacité ❌ (H0 non rejetée)")
```

```python
# EXERCICE 7

import numpy as np

# Création des deux tableaux
A = np.array([[5, 2, 9],
              [1, 7, 3]])

B = np.array([[3, 4, 8],
              [2, 6, 5]])

# Comparaison élément par élément
result = A > B

# Affichage du résultat
print("Tableau A :")
print(A)

print("\nTableau B :")
print(B)

print("\nA > B (résultat booléen) :")
print(result)
```

```python
# EXERCICE 8

import numpy as np
import pandas as pd

# Génération des dates 2023
dates = pd.date_range(start="2023-01-01", end="2023-12-31", freq="D")

# Données aléatoires
values = np.random.randint(0, 100, len(dates))

# DataFrame
df = pd.DataFrame({"Date": dates, "Valeur": values})
df.set_index("Date", inplace=True)

# Découpage
Q1 = df["2023-01-01":"2023-03-31"]
Q2 = df["2023-04-01":"2023-06-30"]
Q3 = df["2023-07-01":"2023-09-30"]
Q4 = df["2023-10-01":"2023-12-31"]

print("=== Jan - Mar ===")
print(Q1)

print("\n=== Avr - Juin ===")
print(Q2)

print("\n=== Juil - Sept ===")
print(Q3)

print("\n=== Oct - Déc ===")
print(Q4)
```

```python
# EXERCICE 9

import numpy as np
import pandas as pd

# =========================
# 1. Création d'un tableau NumPy
# =========================
array_np = np.array([
    [10, 20, 30],
    [40, 50, 60],
    [70, 80, 90]
])

print("Tableau NumPy :")
print(array_np)

# =========================
# 2. Conversion NumPy → DataFrame
# =========================
df = pd.DataFrame(array_np, columns=["A", "B", "C"])

print("\nDataFrame Pandas :")
print(df)

# =========================
# 3. Conversion Pandas → NumPy
# =========================
array_back = df.to_numpy()

print("\nRetour en NumPy :")
print(array_back)
```
