```python
# Challenge

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Charger le dataset (adapter le nom du fichier)
df = pd.read_csv("global_power_plant_database.csv")

# Aperçu
print(df.head())
print(df.info())
```

```python
print(df.isnull().sum())
```

```python
# Exemple : remplacer valeurs manquantes numériques par la médiane
numeric_cols = df.select_dtypes(include=[np.number]).columns

df[numeric_cols] = df[numeric_cols].apply(lambda x: x.fillna(x.median()))
```

```python
df["capacity_mw"] = df["capacity_mw"].astype(float)

capacity_np = df["capacity_mw"].to_numpy()
```

```python
print(df.describe())
```

```python
print("Moyenne capacité :", np.mean(capacity_np))
print("Médiane capacité :", np.median(capacity_np))
print("Écart-type :", np.std(capacity_np))
```

```python
plants_by_country = df["country_long"].value_counts()

print(plants_by_country.head(10))
```

```python
fuel_counts = df["primary_fuel"].value_counts()

print(fuel_counts)
```

```python
fuel_groups = df.groupby("primary_fuel")["capacity_mw"]

fuel_mean = fuel_groups.mean()
fuel_std = fuel_groups.std()

print(fuel_mean)
```

```python
fuel_groups = df.groupby("primary_fuel")["capacity_mw"]

fuel_mean = fuel_groups.mean()
fuel_std = fuel_groups.std()

print(fuel_mean)
```

```python
coal = df[df["primary_fuel"] == "Coal"]["capacity_mw"].to_numpy()
solar = df[df["primary_fuel"] == "Solar"]["capacity_mw"].to_numpy()

print("Coal mean:", np.mean(coal))
print("Solar mean:", np.mean(solar))
```

```python
from scipy import stats

t_stat, p_value = stats.ttest_ind(coal, solar, equal_var=False)

print("t-stat:", t_stat)
print("p-value:", p_value)

if p_value < 0.05:
    print("Différence significative entre Coal et Solar ✅")
else:
    print("Pas de différence significative ❌")
```

```python
df["commissioning_year"] = pd.to_numeric(df["commissioning_year"], errors="coerce")

yearly = df.groupby("commissioning_year")["capacity_mw"].sum()

print(yearly.head())
```

```python
plt.figure(figsize=(10,5))
plt.plot(yearly.index, yearly.values)

plt.title("Évolution de la capacité électrique mondiale")
plt.xlabel("Année")
plt.ylabel("Capacité MW")

plt.grid()
plt.show()
```
