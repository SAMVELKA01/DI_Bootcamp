Exercice 2 : Sélection des caractéristiques et choix du modèle pour la prédiction des défauts de paiement de prêts
1. Caractéristiques les plus pertinentes

Pour prédire les défauts de paiement, les variables suivantes sont particulièrement importantes :

# Caractéristique	                               |                               Importance

# ApplicantIncome (revenu du demandeur)            |       Un revenu élevé augmente généralement la capacité de remboursement.

# CoapplicantIncome (revenu du co-demandeur)	   |       Fournit une source de revenu supplémentaire qui réduit le risque de défaut.

# LoanAmount (montant du prêt)	                   | Les prêts plus importants sont souvent plus difficiles à rembourser.

# Loan_Amount_Term (durée du prêt)                 |	Influence le montant des mensualités et donc la capacité de remboursement.

# Credit_History (historique de crédit)            |	L'un des meilleurs indicateurs du comportement financier passé du client.

# Education (niveau d'éducation)                   |	Peut être lié à la stabilité professionnelle et au niveau de revenu.

# Self_Employed (travailleur indépendant ou non)   |	Les revenus des indépendants peuvent être plus variables.

# Property_Area (zone géographique)                |	Certaines régions présentent des profils économiques différents.

# Married (situation matrimoniale)                 |	Peut refléter une stabilité financière plus importante dans certains cas.

# Dependents (nombre de personnes à charge)	       | Plus de personnes à charge peuvent augmenter les contraintes financières.
# Gender (genre)                                   |	Peut être analysé mais son impact prédictif est généralement faible et doit être utilisé avec prudence pour éviter les biais.

2. Justification du choix des caractéristiques
# a- Historique de crédit (Credit_History)

Cette variable est souvent la plus importante. Un client ayant déjà remboursé correctement ses emprunts présente généralement un risque plus faible qu'un client ayant connu des retards ou défauts de paiement.

# b- Revenus (ApplicantIncome et CoapplicantIncome)

Les revenus permettent d'évaluer la capacité financière du demandeur à effectuer ses remboursements mensuels.

# c- Montant du prêt (LoanAmount)

Plus le montant emprunté est élevé, plus la charge financière est importante, ce qui peut accroître le risque de défaut.

# d- Durée du prêt (Loan_Amount_Term)

Une durée plus longue réduit souvent les mensualités mais augmente la période d'exposition au risque.

# e- Nombre de personnes à charge (Dependents)

Les charges familiales influencent directement la capacité de remboursement du ménage.

# f- Niveau d'éducation (Education)

Le niveau d'éducation peut être associé à de meilleures opportunités d'emploi et à une stabilité financière plus élevée.

# g- Situation professionnelle (Self_Employed)

Les travailleurs indépendants peuvent avoir des revenus moins réguliers, ce qui peut affecter leur capacité à rembourser un prêt.

# h- Zone géographique (Property_Area)

Les conditions économiques varient selon les régions, ce qui peut influencer les revenus, l'emploi et donc le risque de défaut.

3. Choix du modèle

Comme l'objectif est de prédire deux résultats possibles :

0 : Pas de défaut de paiement
1 : Défaut de paiement

il s'agit d'un problème de classification binaire.

Les modèles adaptés sont :

Régression Logistique
Simple à interpréter.
Bonne base de référence.
Arbre de Décision
Facile à comprendre.
Capture les relations non linéaires.
Random Forest
Plus robuste que l'arbre de décision.
Réduit le risque de surapprentissage.
Généralement très performant sur ce type de données.
Gradient Boosting / XGBoost
Excellentes performances prédictives.
Très utilisé dans le secteur bancaire.

Conclusion

Les variables Credit_History, ApplicantIncome, CoapplicantIncome, LoanAmount et Loan_Amount_Term sont les plus importantes pour prédire les défauts de paiement. Le problème étant une classification binaire, des modèles comme la Régression Logistique, les Arbres de Décision ou le Random Forest constituent d'excellents choix, avec une préférence pour le Random Forest ou XGBoost lorsque l'objectif principal est la précision des prédictions.

Exercice 5 : Concevoir une stratégie d'évaluation pour différents modèles d'apprentissage automatique
Introduction

L'évaluation des modèles d'apprentissage automatique est une étape essentielle pour mesurer leur efficacité et leur capacité à généraliser sur de nouvelles données. Les méthodes d'évaluation varient selon le type d'apprentissage utilisé : supervisé, non supervisé ou par renforcement.

# 1. Modèle d'apprentissage supervisé : Classification des défauts de paiement de prêts
Modèle choisi

Random Forest Classifier

Stratégie d'évaluation
# a- Séparation des données

Le jeu de données est divisé en :

80 % pour l'entraînement
20 % pour les tests
# b- Validation croisée (Cross Validation)

La validation croisée (K-Fold) consiste à diviser les données en plusieurs sous-ensembles afin d'entraîner et de tester le modèle plusieurs fois. Cette méthode fournit une estimation plus fiable des performances.

# c- Métriques d'évaluation

Accuracy (Exactitude)

Mesure la proportion de prédictions correctes.
Utile lorsque les classes sont équilibrées.

Precision (Précision)

Mesure la proportion de prédictions positives correctes.
Importante pour limiter les faux positifs.

Recall (Rappel)

Mesure la capacité à détecter les vrais cas positifs.
Crucial dans la détection des défauts de paiement.

F1-Score

Combine précision et rappel.
Adapté aux jeux de données déséquilibrés.

Courbe ROC et AUC

Évaluent la capacité du modèle à distinguer les différentes classes.
Une AUC proche de 1 indique un excellent modèle.
Difficultés et limites
Déséquilibre possible entre les classes.
Risque de surapprentissage.
Une forte exactitude peut être trompeuse lorsque les données sont déséquilibrées.
# 2. Modèle d'apprentissage non supervisé : Regroupement de livres par catégories
Modèle choisi

K-Means Clustering

Stratégie d'évaluation

Contrairement à l'apprentissage supervisé, il n'existe pas de valeurs réelles permettant de comparer directement les résultats.

# a- Score de silhouette

Le score de silhouette mesure la qualité du regroupement.

Valeur proche de 1 : clusters bien séparés.
Valeur proche de 0 : chevauchement entre clusters.
Valeur négative : mauvaise classification.
# b- Méthode du coude (Elbow Method)

Cette méthode permet de déterminer le nombre optimal de clusters.

Le point où la diminution de l'erreur devient moins significative représente généralement le nombre idéal de groupes.

# c- Validation des clusters

On peut analyser :

La cohérence interne des groupes.
La séparation entre les groupes.
L'interprétabilité des résultats obtenus.
Difficultés et limites
Absence de vérité terrain (labels réels).
Résultats sensibles au choix du nombre de clusters.
Les métriques ne garantissent pas toujours une interprétation métier pertinente.
# 3. Modèle d'apprentissage par renforcement : Navigation d'un robot dans un labyrinthe
Modèle choisi

Q-Learning

Stratégie d'évaluation
# a- Récompense cumulative

La récompense totale obtenue au cours des épisodes mesure directement les performances du robot.

Une augmentation progressive indique un apprentissage efficace.

# b- Taux de réussite

Mesure la proportion d'épisodes où le robot atteint la sortie du labyrinthe.

# c- Nombre moyen d'étapes

Permet d'évaluer l'efficacité du chemin trouvé.

Moins d'étapes signifie généralement une meilleure stratégie.

# d- Convergence

On observe si les performances se stabilisent après un certain nombre d'épisodes.

Une convergence indique que le robot a appris une politique efficace.

# e- Équilibre exploration/exploitation

Le robot doit :

Explorer de nouveaux chemins.
Exploiter les meilleures stratégies découvertes.

Un bon équilibre améliore l'apprentissage.

Difficultés et limites
Temps d'entraînement souvent long.
Sensibilité au système de récompenses.
Risque de convergence vers une solution sous-optimale.
Évaluation parfois coûteuse en ressources de calcul.
Conclusion
