```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.decomposition import PCA

# Charger le dataset
df = pd.read_csv("ds_salaries.csv")

# -----------------------------
# 1. Normalisation Min-Max
# -----------------------------
scaler = MinMaxScaler()

df["salary_normalized"] = scaler.fit_transform(df[["salary"]])

print(df[["salary", "salary_normalized"]].head())

# -----------------------------
# 2. Réduction de dimension (PCA)
# -----------------------------

# Sélection des colonnes numériques
numeric_cols = df.select_dtypes(include="number")

# PCA : réduction à 2 composantes
pca = PCA(n_components=2)

pca_result = pca.fit_transform(numeric_cols)

# Ajouter les composantes au dataset
df["PCA1"] = pca_result[:, 0]
df["PCA2"] = pca_result[:, 1]

print("\nVariance expliquée :")
print(pca.explained_variance_ratio_)

# -----------------------------
# 3. Agrégation des données
# -----------------------------

salary_stats = df.groupby("experience_level")["salary"].agg(["mean", "median"])

print("\nSalaire moyen et médian par niveau d'expérience :")
print(salary_stats)
```
