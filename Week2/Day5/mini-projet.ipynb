```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Chargement
df = pd.read_excel("Mini-project/US Superstore data.xls")

# Vérifications
print(df.shape)
print(df.info())
print(df.isnull().sum())

# Suppression des doublons
df = df.drop_duplicates()

# Conversion des dates
df["Order Date"] = pd.to_datetime(df["Order Date"])
df["Ship Date"] = pd.to_datetime(df["Ship Date"])
```

```python
state_sales = (
    df.groupby("State")["Sales"]
    .sum()
    .sort_values(ascending=False)
)

plt.figure(figsize=(12,6))
sns.barplot(
    x=state_sales.head(10).values,
    y=state_sales.head(10).index
)

plt.title("Top 10 États par chiffre d'affaires")
plt.xlabel("Ventes")
plt.ylabel("État")
plt.show()
```

```python
comparison = (
    df[df["State"].isin(["California","New York"])]
    .groupby("State")[["Sales","Profit"]]
    .sum()
)

print(comparison)
```

```python
ny_customers = (
    df[df["State"]=="New York"]
    .groupby("Customer Name")["Sales"]
    .sum()
    .sort_values(ascending=False)
)

print(ny_customers.head(10))
```

```python
state_profitability = (
    df.groupby("State")
    .agg({
        "Sales":"sum",
        "Profit":"sum"
    })
)

state_profitability["Profit Margin"] = (
    state_profitability["Profit"]
    /
    state_profitability["Sales"]
)*100

state_profitability.sort_values(
    "Profit Margin",
    ascending=False
).head()
```

```python
customer_profit = (
    df.groupby("Customer Name")["Profit"]
    .sum()
    .sort_values(ascending=False)
)

cum_profit = (
    customer_profit.cumsum()
    /
    customer_profit.sum()
)

pareto_df = pd.DataFrame({
    "Profit":customer_profit,
    "Cumulative":cum_profit
})

pareto_df.head()
```

```python
city_sales = (
    df.groupby("City")["Sales"]
    .sum()
    .sort_values(ascending=False)
    .head(20)
)

plt.figure(figsize=(12,8))

sns.barplot(
    x=city_sales.values,
    y=city_sales.index
)

plt.title("Top 20 villes par ventes")
plt.show()
```

```python
city_profit = (
    df.groupby("City")["Profit"]
    .sum()
    .sort_values(ascending=False)
    .head(20)
)

plt.figure(figsize=(12,8))

sns.barplot(
    x=city_profit.values,
    y=city_profit.index
)

plt.title("Top 20 villes par bénéfices")
plt.show()
```

```python
city_analysis = (
    df.groupby("City")
    .agg({
        "Sales":"sum",
        "Profit":"sum"
    })
)

city_analysis["Profit Margin"] = (
    city_analysis["Profit"]
    /
    city_analysis["Sales"]
)*100

city_analysis.sort_values(
    "Profit Margin",
    ascending=False
).head(20)
```

```python
top_customers = (
    df.groupby("Customer Name")["Sales"]
    .sum()
    .sort_values(ascending=False)
    .head(20)
)

plt.figure(figsize=(12,8))

sns.barplot(
    x=top_customers.values,
    y=top_customers.index
)

plt.title("Top 20 clients par ventes")
plt.show()
```

```python
customer_sales = (
    df.groupby("Customer Name")["Sales"]
    .sum()
    .sort_values(ascending=False)
)

cum_sales = (
    customer_sales.cumsum()
    /
    customer_sales.sum()
)

plt.figure(figsize=(10,6))

plt.plot(
    range(len(cum_sales)),
    cum_sales
)

plt.axhline(
    y=0.8,
    color="red",
    linestyle="--"
)

plt.title(
    "Courbe cumulative des ventes par client"
)

plt.xlabel("Clients")
plt.ylabel("Pourcentage cumulé")

plt.show()
```

1. California et New York génèrent la plus grande partie du chiffre d'affaires.

2. Certaines villes génèrent beaucoup de ventes mais peu de bénéfices, indiquant une faible rentabilité.

3. Un petit nombre de clients contribue à la majorité des ventes et profits (principe de Pareto).

4. Les campagnes marketing doivent prioritairement cibler les États et villes les plus rentables.

5. Les clients premium doivent faire l'objet de programmes de fidélisation spécifiques.
