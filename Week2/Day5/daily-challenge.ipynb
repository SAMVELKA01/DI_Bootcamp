```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Chargement
df = pd.read_excel("Mini-project/US Superstore data.xls")

# Informations générales
print(df.shape)
print(df.info())

# Valeurs manquantes
print(df.isnull().sum())

# Suppression des doublons
df = df.drop_duplicates()

# Conversion des dates
df["Order Date"] = pd.to_datetime(df["Order Date"])
df["Ship Date"] = pd.to_datetime(df["Ship Date"])

# Création de l'année
df["Order Year"] = df["Order Date"].dt.year
```

```python
from ipywidgets import interact

annual_sales = df.groupby("Order Year")["Sales"].sum()

def sales_trend():
    plt.figure(figsize=(10,5))
    plt.plot(
        annual_sales.index,
        annual_sales.values,
        marker="o",
        linewidth=2
    )

    plt.title("Évolution des ventes par année")
    plt.xlabel("Année")
    plt.ylabel("Ventes")
    plt.grid(True)
    plt.show()

interact(sales_trend)
```

```python
country_sales = df.groupby("Country")["Sales"].sum()

plt.figure(figsize=(8,5))

plt.bar(
    country_sales.index,
    country_sales.values
)

plt.title("Répartition des ventes par pays")
plt.xlabel("Pays")
plt.ylabel("Ventes")

plt.show()
```

```python
top_products = (
    df.groupby("Product Name")["Sales"]
      .sum()
      .sort_values(ascending=False)
      .head(10)
)

plt.figure(figsize=(12,6))

sns.barplot(
    x=top_products.values,
    y=top_products.index,
    palette="viridis"
)

plt.title("Top 10 produits les plus vendus")
plt.xlabel("Ventes")
plt.ylabel("Produit")

plt.show()
```

```python
plt.figure(figsize=(10,6))

sns.scatterplot(
    data=df,
    x="Discount",
    y="Profit",
    hue="Category",
    alpha=0.7
)

plt.title("Relation entre Profit et Discount")
plt.xlabel("Discount")
plt.ylabel("Profit")

plt.show()
```

# Analyse comparative : Matplotlib vs Seaborn

## Matplotlib

Avantages :

* Contrôle très précis des graphiques.
* Intégration simple avec les widgets interactifs.
* Personnalisation avancée.

Inconvénients :

* Nécessite davantage de code.
* Style visuel moins moderne par défaut.

## Seaborn

Avantages :

* Graphiques élégants avec peu de code.
* Excellente gestion des données catégorielles.
* Palettes de couleurs professionnelles.

Inconvénients :

* Moins flexible pour certaines personnalisations avancées.
* Dépend de Matplotlib en arrière-plan.

## Conclusion

Matplotlib est particulièrement adapté aux visualisations interactives et aux personnalisations avancées. Seaborn est plus efficace pour produire rapidement des graphiques professionnels destinés à l'analyse exploratoire et à la présentation des résultats.


# Observations

* Les ventes présentent une tendance globale à la hausse au fil des années.
* Certains produits génèrent une part importante du chiffre d'affaires.
* Une augmentation des remises importantes est souvent associée à une diminution des profits.
* Seaborn facilite l'analyse visuelle des relations entre variables.
* Matplotlib offre davantage de flexibilité pour créer des visualisations interactives.
