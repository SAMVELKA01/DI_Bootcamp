Pourquoi la visualisation des données est importante dans l’analyse des données

La visualisation des données est importante car elle permet de transformer des données brutes en informations faciles à comprendre. Grâce aux graphiques et aux diagrammes, on peut :

repérer rapidement des tendances ;
identifier des anomalies ou des valeurs inhabituelles ;
comparer différentes données ;
mieux comprendre de grands ensembles de données ;
faciliter la prise de décision ;
communiquer clairement les résultats d’une analyse.

Sans visualisation, l’interprétation de grandes quantités de données serait beaucoup plus difficile et moins efficace.

Utilité d’un graphique linéaire en visualisation de données

Un graphique linéaire sert principalement à montrer l’évolution d’une donnée au fil du temps. Les points représentant les valeurs sont reliés par des lignes pour rendre les changements plus visibles.

Il est utile pour :

observer des tendances (hausse ou baisse) ;
suivre des variations dans le temps ;
comparer plusieurs séries de données ;
détecter des pics ou des chutes importantes.

Exemple : suivre l’évolution des ventes mensuelles d’une entreprise ou la température sur plusieurs jours.

```python
# EXERCICE 2
import matplotlib.pyplot as plt

jours = ["Lun", "Mar", "Mer", "Jeu", "Ven", "Sam", "Dim"]
temperatures = [72, 74, 76, 80, 82, 78, 75]

plt.plot(jours, temperatures, marker='o')

plt.title("Variation de la température sur une semaine")
plt.xlabel("Jour")
plt.ylabel("Température (°F)")

plt.show()
```

```python
# EXERCICE 5
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Exemple de données
data = {
    "Choisissez votre sexe": ["Homme", "Femme", "Homme", "Femme", "Homme", "Femme"],
    "Souffrez-vous d’anxiété ?": ["Oui", "Oui", "Non", "Oui", "Non", "Non"]
}

# Création du DataFrame
df = pd.DataFrame(data)

# Création du graphique à barres
sns.countplot(
    data=df,
    x="Choisissez votre sexe",
    hue="Souffrez-vous d’anxiété ?",
    palette="Set2"
)

# Personnalisation du graphique
plt.title("Comparaison des niveaux d’anxiété selon le sexe")
plt.xlabel("Sexe")
plt.ylabel("Nombre d'étudiants")

# Affichage du graphique
plt.show()
```

```python
# EXERCICE 6
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Importation du dataset
df = pd.read_csv("student_mental_health.csv")

# Création du diagramme à barres
plt.figure(figsize=(8, 5))

sns.countplot(
    data=df,
    x="Choose your gender",
    hue="Do you have Anxiety?",
    palette="Set2"
)

# Personnalisation du graphique
plt.title("Comparaison des niveaux d'anxiété selon le sexe")
plt.xlabel("Sexe")
plt.ylabel("Nombre d'étudiants")
plt.legend(title="Anxiété")

# Affichage
plt.show()
```
