```python
# Défi quotidien : Analyse des accidents d’avion et des décès jusqu’en 2023

## 1. Introduction

# Dans ce projet, nous allons analyser le jeu de données **“Airplane Crashes and Fatalities Until 2023”** en utilisant **Python**, **Pandas**, **NumPy**, **SciPy**, **Matplotlib** et **Seaborn**.

# L’objectif est de :

# * Nettoyer et préparer les données
# * Explorer les tendances des accidents d’avion
# * Réaliser des analyses statistiques
# * Visualiser les résultats
# * Tirer des conclusions pertinentes
```

```python
# 2. Importation des bibliothèques


import pandas as pd
import numpy as np
from scipy import stats

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 3. Chargement du jeu de données


# Charger le fichier CSV
df = pd.read_csv(
    "Airplane_Crashes_and_Fatalities_2023.csv",
    encoding="latin1"
)
# Aperçu des données
print(df.head())

# Informations générales
print(df.info())
```

```python
# 4. Nettoyage et prétraitement des données

## Vérification des valeurs manquantes

print(df.isnull().sum())

## Suppression ou remplacement des valeurs manquantes


# Remplacer les valeurs manquantes numériques par 0
df['Fatalities'] = df['Fatalities'].fillna(0)

# Supprimer les lignes sans date
df = df.dropna(subset=['Date'])


## Conversion des dates


df['Date'] = pd.to_datetime(df['Date'])

# Extraction de l’année
df['Year'] = df['Date'].dt.year


## Création du taux de survie


# Exemple :
# Aboard = nombre total de personnes à bord
# Fatalities = décès

df['Survival_Rate'] = (
    (df['Aboard'] - df['Fatalities']) / df['Aboard']
) * 100

# Gestion des divisions par zéro
df['Survival_Rate'] = df['Survival_Rate'].replace([np.inf, -np.inf], np.nan)
```

```python
# 5. Analyse exploratoire des données (EDA)

## Nombre total d’accidents


print("Nombre total d'accidents :", len(df))


## Nombre total de décès


print("Nombre total de décès :", df['Fatalities'].sum())


## Moyenne des décès


print("Moyenne des décès :", df['Fatalities'].mean())


## Accidents par année


accidents_per_year = df.groupby('Year').size()

print(accidents_per_year.head())
```

```python
# 6. Analyse statistique avec SciPy

## Statistiques descriptives


mean_fatalities = np.mean(df['Fatalities'])
median_fatalities = np.median(df['Fatalities'])
std_fatalities = np.std(df['Fatalities'])

print("Moyenne :", mean_fatalities)
print("Médiane :", median_fatalities)
print("Écart type :", std_fatalities)




## Distribution des décès


print(stats.describe(df['Fatalities']))
```

```python
# 7. Test d’hypothèse

## Comparaison des décès entre deux périodes


before_1980 = df[df['Year'] < 1980]['Fatalities']
after_1980 = df[df['Year'] >= 1980]['Fatalities']

# Test t
t_stat, p_value = stats.ttest_ind(before_1980, after_1980)

print("Statistique t :", t_stat)
print("Valeur p :", p_value)


## Interprétation


if p_value < 0.05:
    print("Différence statistiquement significative.")
else:
    print("Aucune différence significative.")
```

```python
# 8. Visualisations

## Style des graphiques


sns.set(style="whitegrid")


## Graphique chronologique des accidents


plt.figure(figsize=(12,6))

accidents_per_year.plot()

plt.title("Nombre d'accidents d'avion par année")
plt.xlabel("Année")
plt.ylabel("Nombre d'accidents")

plt.show()

## Histogramme des décès


plt.figure(figsize=(10,6))

sns.histplot(df['Fatalities'], bins=30)

plt.title("Distribution des décès")
plt.xlabel("Nombre de décès")
plt.ylabel("Fréquence")

plt.show()


## Diagramme des accidents par région


top_regions = df['Location'].value_counts().head(10)

plt.figure(figsize=(12,6))

sns.barplot(
    x=top_regions.values,
    y=top_regions.index
)

plt.title("Top 10 des régions avec le plus d'accidents")
plt.xlabel("Nombre d'accidents")
plt.ylabel("Région")

plt.show()
```

```python
# 9. Analyse des résultats

## Observations principales

### 1. Évolution des accidents

# Le nombre d’accidents varie selon les décennies. Certaines périodes montrent une diminution progressive des incidents grâce aux progrès technologiques et aux améliorations des normes de sécurité aérienne.

### 2. Distribution des décès

# La majorité des accidents comportent relativement peu de décès, mais quelques catastrophes majeures provoquent un nombre extrêmement élevé de victimes.

### 3. Taux de survie

# Les taux de survie tendent à s’améliorer dans les périodes récentes grâce :

# * aux avancées technologiques,
# * aux procédures de sécurité,
# * à une meilleure formation des équipages.

### 4. Résultat du test statistique

# Le test t permet de vérifier si les décès moyens diffèrent significativement entre deux périodes historiques.
```
