```python
# EXERCICE 4

import numpy as np
from scipy import stats

# Génération des données aléatoires
data1 = np.random.normal(50, 10, 100)
data2 = np.random.normal(60, 10, 100)

# Test t indépendant
t_stat, p_value = stats.ttest_ind(data1, data2)

# Affichage des résultats
print("Statistique t :", t_stat)
print("Valeur p :", p_value)

# Interprétation
if p_value < 0.05:
    print("Les deux groupes sont significativement différents.")
else:
    print("Aucune différence significative entre les deux groupes.")
```

```python
# EXERCICE 5

import numpy as np
from scipy.stats import linregress

# Données
house_sizes = np.array([50, 70, 80, 100, 120])
house_prices = np.array([150000, 200000, 210000, 250000, 280000])

# Régression linéaire
result = linregress(house_sizes, house_prices)

# Paramètres de la droite
slope = result.slope
intercept = result.intercept

print("Pente :", slope)
print("Ordonnée à l'origine :", intercept)

# Prédiction pour une maison de 90 m²
size = 90
predicted_price = slope * size + intercept

print(f"Prix estimé pour une maison de {size} m² : {predicted_price:.2f}")

# Interprétation
print(f"""
Interprétation :
La pente indique qu'à chaque augmentation de 1 m²,
le prix de la maison augmente en moyenne de {slope:.2f} unités monétaires.
""")
```

```python
#  EXERCICE 6

from scipy.stats import f_oneway

# Données
fertilizer_1 = [5, 6, 7, 6, 5]
fertilizer_2 = [7, 8, 7, 9, 8]
fertilizer_3 = [4, 5, 4, 3, 4]

# Test ANOVA
f_stat, p_value = f_oneway(fertilizer_1, fertilizer_2, fertilizer_3)

# Résultats
print("Valeur F :", f_stat)
print("Valeur p :", p_value)

# Interprétation
if p_value < 0.05:
    print("Les engrais ont des effets significativement différents sur la croissance des plantes.")
else:
    print("Aucune différence significative entre les engrais.")

print("""
Si la valeur p était supérieure à 0,05 :
cela signifierait qu'il n'y a pas suffisamment de preuves
pour conclure que les engrais ont des effets différents.
Les différences observées pourraient être dues au hasard.
""")
```

```python
# EXERCICE 7

from scipy.stats import binom

# Paramètres de la distribution binomiale
n = 10      # nombre de lancers
p = 0.5     # probabilité d'obtenir face
k = 5       # nombre exact de faces

# Calcul de la probabilité
probability = binom.pmf(k, n, p)

# Affichage du résultat
print(f"Probabilité d'obtenir exactement {k} faces en {n} lancers :")
print(probability)
```

```python
# EXERCICE 8

import pandas as pd
from scipy.stats import pearsonr, spearmanr

# Création du DataFrame
data = pd.DataFrame({
    'age': [23, 25, 30, 35, 40],
    'income': [35000, 40000, 50000, 60000, 70000]
})

# Variables
age = data['age']
income = data['income']

# Corrélation de Pearson
pearson_corr, pearson_p = pearsonr(age, income)

# Corrélation de Spearman
spearman_corr, spearman_p = spearmanr(age, income)

# Affichage des résultats
print("Corrélation de Pearson :", pearson_corr)
print("Valeur p (Pearson) :", pearson_p)

print("\nCorrélation de Spearman :", spearman_corr)
print("Valeur p (Spearman) :", spearman_p)
```
